```python
%%writefile app.py
import streamlit as st
import pandas as pd
import os
from openai import OpenAI

client = OpenAI(api_key=os.getenv("OPENAI_API_KEY"))

st.set_page_config(page_title="Smart Data Insight Agent", page_icon="ðŸ¤–", layout="wide")

st.markdown("<h1 style='text-align:center; color:#0078D7;'>ðŸ“Š Smart Data Insight Agent</h1>", unsafe_allow_html=True)
st.write("Upload a CSV file and ask natural-language questions â€” powered by OpenAI GPT.")

uploaded_file = st.file_uploader("ðŸ“‚ Upload a CSV file", type=["csv"])

if uploaded_file:
    df = pd.read_csv(uploaded_file)
    st.subheader("ðŸ§¾ Data Preview")
    st.dataframe(df.head())

    query = st.text_input("ðŸ’¬ Ask a question about your data:")

    if st.button("Get Insight") and query.strip():
        with st.spinner("ðŸ¤– Thinking..."):
            try:
                # Convert small sample of dataset to context
                data_sample = df.head(10).to_csv(index=False)
                prompt = f"""
                You are a data analyst AI. Analyze the following dataset and answer the question.
                Dataset (CSV Sample):
                {data_sample}

                Question: {query}
                Answer in simple English.
                """

                response = client.chat.completions.create(
                    model="gpt-3.5-turbo",
                    messages=[{"role": "user", "content": prompt}]
                )
                answer = response.choices[0].message.content
                st.success("âœ… Insight:")
                st.write(answer)
            except Exception as e:
                st.error(f"âš ï¸ Error: {e}")
else:
    st.info("ðŸ‘† Upload a CSV file to begin.")
```
